```python
import pandas as pd
import pickle
import numpy as np
```

```python
with open('puzzles.pickle','rb') as rf:
    db = pickle.load(rf)
```

```python
def search(category, word_lengths):
    return db[(db.category==category) & (db.word_lengths==word_lengths)]
```
